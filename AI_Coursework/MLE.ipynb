```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Step 1: Parameters
theta_true = 0.8   # True probability of heads
n = 5              # Number of tosses
theta_vals = np.linspace(0.01, 0.99, 100)  # Range of theta for plotting

# Step 2: Likelihood, Log-Likelihood, Negative Log-Likelihood functions
def likelihood(theta, k, n):
    return (theta**k) * ((1-theta)**(n-k))

def log_likelihood(theta, k, n):
    return k*np.log(theta) + (n-k)*np.log(1-theta)

def neg_log_likelihood(theta, k, n):
    return -log_likelihood(theta, k, n)

# Step 3: Plotting for each case (k=0 to 5 heads)
fig, axes = plt.subplots(6, 3, figsize=(15, 20))

for k in range(6):
    # Compute values
    L = [likelihood(t, k, n) for t in theta_vals]
    logL = [log_likelihood(t, k, n) for t in theta_vals]
    nlogL = [neg_log_likelihood(t, k, n) for t in theta_vals]

    # Likelihood
    axes[k, 0].plot(theta_vals, L, label=f"Heads={k}, Tails={n-k}")
    axes[k, 0].set_title(f"Likelihood (k={k})")
    axes[k, 0].set_xlabel("")
    axes[k, 0].set_ylabel("L()")
    axes[k, 0].legend()

    # Log-Likelihood
    axes[k, 1].plot(theta_vals, logL, color="green")
    axes[k, 1].set_title(f"Log-Likelihood (k={k})")
    axes[k, 1].set_xlabel("")
    axes[k, 1].set_ylabel("log L()")

    # Negative Log-Likelihood
    axes[k, 2].plot(theta_vals, nlogL, color="red")
    axes[k, 2].set_title(f"Neg Log-Likelihood (k={k})")
    axes[k, 2].set_xlabel("")
    axes[k, 2].set_ylabel("-log L()")

plt.tight_layout()
plt.show()
```
